Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
import re
import string
```

Importing Datasets

```python
fake = pd.read_csv("/content/Fake.csv")
true = pd.read_csv("/content/True.csv")
```

Combining Datasets

```python
news = pd.concat([fake, true], axis=0)
```

```python
news.head()
```

```python
news.isnull().sum()
```

Dropping Unnecessaries Columns

```python
news = news.drop(['title','subject','date'], axis=1)
```

```python
news.head()
```

Shuffling the rows


```python
news = news.sample(frac=1)
```

```python
news.head()
```

Indexing

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

Dropping the 'Index' column

```python
news.drop(['index'], axis=1, inplace=True)
```

```python
news.head()
```

Removing URLs, HTML tags, punctuation, digits and newline characters from the text

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+','',text)
    text = re.sub(r'<.*?>','',text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d','',text)
    text = re.sub(r'\n','',text)

    return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
x
```

```python
y
```

Traning and Testing the data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

Logistic Regression

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

Decision Tree Classifier

```python
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

Gradient Boosting Classifier

```python
GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train, y_train)
```

```python
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

Random Forest Classifier

```python
RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

Function for Testing the input news

```python
def output_label(n):
    if  n == 0:
        return "Fake News"
    elif n == 1:
        return "Not a Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_lr = LR.predict(new_xv_test)
    pred_dt = DT.predict(new_xv_test)
    pred_gb = GB.predict(new_xv_test)
    pred_rf = RF.predict(new_xv_test)

    return "LR Prediction: {}   DT Prediction: {}   RF Prediction: {}    GB Prediction: {}".format(
    output_label(pred_lr[0]), output_label(pred_dt[0]), output_label(pred_rf[0]), output_label(pred_gb[0]))
```

Input News Article

```python
news_article = str(input())
```

Checking the Article is Fake or not

```python
manual_testing(news_article)
```
